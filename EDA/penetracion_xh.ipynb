En el siguiente notebook, se va a trabajar con los datos con el fin de poder sacar el máximo provecho de los datos de una forma más ordenada, detectando los valores faltantes, los datos duplicados y atípicos para que no afecte de manera negativa al análisis. 


```python
# Se importan las librerías necesarias

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
# Se cargan los datos de la hoja llamada Penetracion-hogares 
df_internet_pxh = pd.read_excel('../Data/Internet.xlsx', sheet_name='Penetracion-hogares')
```

```python
df_internet_pxh.head()
```

```python
# Se realiza una búsqueda de los valores faltantes
valores_faltantes = df_internet_pxh.isnull().sum()

# Muestra de los valores faltantes
valores_faltantes = valores_faltantes[valores_faltantes > 0]
valores_faltantes
```

```python
# Verificación de registros duplicados
duplicados = df_internet_pxh.duplicated().sum()
duplicados
```

No se detectan valores nulos ni duplicados, por loque no es necesario ningún proceso de eliminación o reemplazo de valores.

```python
# Configuración del estilo de los gráficos
sns.set(style="whitegrid")

# Creación del diagrama de caja
plt.figure(figsize=(12, 6))
sns.boxplot(x='Provincia', y='Accesos por cada 100 hogares', data=df_internet_pxh)

plt.title('Diagrama de Caja de Accesos por cada 100 Hogares por Provincia')
plt.xlabel('Provincia')
plt.ylabel('Accesos por cada 100 hogares')

# Mostrar el gráfico
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
# Histograma de accesos
plt.figure(figsize=(10, 6))
sns.histplot(df_internet_pxh,x='Provincia', y='Accesos por cada 100 hogares', bins=20, kde=True)
plt.title('Distribución de Accesos por cada 100 Hogares')
plt.xlabel('Accesos por cada 100 Hogares')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()

# Boxplot para visualizar la dispersión
plt.figure(figsize=(12, 6))
sns.boxplot(x='Provincia', y='Accesos por cada 100 hogares', data=df_internet_pxh)
plt.title('Accesos por cada 100 Hogares')
plt.xlabel('Accesos por cada 100 Hogares')
plt.xticks(rotation=45)
plt.show()
```

```python
# Tendencia de accesos a lo largo del tiempo por provincia
plt.figure(figsize=(14, 8))
sns.lineplot(data=df_internet_pxh, x='Año', y='Accesos por cada 100 hogares', hue='Provincia')
plt.title('Tendencia de Accesos a lo Largo del Tiempo por Provincia')
plt.xlabel('Año')
plt.ylabel('Accesos por cada 100 Hogares')
plt.legend(title='Provincia', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
